## Weather

```python
import pandas as pd
import os

# 1. Define Paths
input_path = '/content/drive/MyDrive/2025Fall_LLM/Final/cleaned_weather.csv'
output_path = '/content/drive/MyDrive/2025Fall_LLM/Final/weather_20min.csv'

# 2. Load the Dataset
# We assume the first column is the Date/Time.
# If 'date' is not the first column, change index_col=0 to the correct name/index.
df = pd.read_csv(input_path)

# Quick check of the column names to find the Date column
print(f"Original Columns: {df.columns.tolist()}")

# Attempt to identify the date column automatically (usually the first one)
date_col = df.columns[0]
print(f"Assuming '{date_col}' is the datetime column.")

# 3. Convert to Datetime Objects
df[date_col] = pd.to_datetime(df[date_col])

# 4. Set Index and Resample
df = df.set_index(date_col)

# '20T' stands for 20 Minutes in Pandas
# .mean() averages the values in that 20-minute window
df_resampled = df.resample('20min').mean()

# Drop any rows that might have become NaNs during resampling (rare but possible)
df_resampled.dropna(inplace=True)

# 5. Reset Index (so Date becomes a column again)
df_resampled.reset_index(inplace=True)

# 6. Save and Verify
df_resampled.to_csv(output_path, index=False)

print("-" * 30)
print(f"Original Shape:  {df.shape}")
print(f"Resampled Shape: {df_resampled.shape}")
print(f"Saved to: {output_path}")
print("-" * 30)
```
